1-Create a python class circle, constructed by a radius and two methods that will
compute the area and the perimeter of a circle.

```python

import numpy as np

class Circle:
  def __init__(self, radius):
    self.radius = radius

  def perimeter(self):
    return 2*np.pi*self.radius

  def area(self):
    return np.pi*(self.radius**2)

r = int(input("Enter the radius: "))
circle = Circle(r)
print("Perimeter:", circle.perimeter())
print("Area:", circle.area())
```

3. Write a python program to demonstrate the working of the banking system.
a. Create account class with data members (Account No, Account holder name,
Amount)
b. Crete constructor to initialize data members.
c. Create member functions (Deposit, withdraw, check balance, and display account
details)

```python
class account:
  def __init__(self, Name, Account_No, Deposit):
    self.accountNumber = Account_No
    self.Name = Name
    self.Savings = Deposit

  def Deposit(self, amount):
    self.Savings += amount
    print("-----------------------------------------------")
    print(f"Rs {amount} has been deposited")
    print(f"current savings: {self.Savings}")
    print("-----------------------------------------------")

  def Withdraw(self, amount):
    if (amount > self.Savings):
      print("-----------------------------------------------")
      print("Insufficient savings")
      print("-----------------------------------------------")
    else:
      print("-----------------------------------------------")
      self.Savings -= amount
      print(f"Rs {amount} has been withdrawn")
      print(f"current savings: {self.Savings}")
      print("-----------------------------------------------")

  def check_balance(self):
    print("-----------------------------------------------")
    print(f"current savings: {self.Savings}")
    print("-----------------------------------------------")

  def displayAccount(self):
    print("-----------------------------------------------")
    print("Name:", self.Name)
    print("Account Number", self.accountNumber)
    print(f"current savings: {self.Savings}")
    print("-----------------------------------------------")

name = str(input("Enter your name: "))
account_no = int(input("Enter the account number: "))
deposit = int(input("Enter the initial balance: "))
User_acc = account(name, account_no, deposit)

while True:
  choice = int(input("Select an option:\n1. Deposit\n2. Withdraw\n3. Check Balance\n4. Display Account\n5. Exit\n"))
  if choice == 1:
    amount = int(input("Enter the amount to deposit: "))
    User_acc.Deposit(amount)
  elif choice == 2:
    amount = int(input("Enter the amount to withdraw: "))
    User_acc.Withdraw(amount)
  elif choice == 3:
    User_acc.check_balance()
  elif choice == 4:
    User_acc.displayAccount()
  elif choice == 5:
    break
  else:
    print("enter a valid choice")
```

5. A hospital wants to create a database regarding covid-19 patients. The information to
store include:
a) Name of the patient
b) Date of admission.
c) List of symptoms.
d) Oxygen level.
d) Date of discharge.
Note: If the oxygen level is less than 90, then the patient is put on oxygen support else
moved to the general ward for observation.
Create a class to store the above information. The member function should include a
function to enter patient’s information, check the severity, display the number of
patients required oxygen support, number of patients in the general ward for
observation, and the total number of patients admitted.

```python
class Patient:
  def __init__(self, name, admission_date, symptoms : list, oxy_level, discharge_date):
    self.name = name
    self.admission_date = admission_date
    self.symptoms = symptoms
    self.oxy_level = oxy_level
    self.discharge_date = discharge_date

class PatientData:
  def __init__(self):
    self.oxygen_support = []
    self.general_ward = []

  def addPatient(self, patient):
    if (patient.oxy_level < 90):
      self.oxygen_support.append(patient)
    else:
      self.general_ward.append(patient)

  def numberOxygen(self):
    print(f"The number of patients needing oxygen support is {len(self.oxygen_support)}")

  def numberGeneral(self):
    print(f"The number of patients in general ward is {len(self.general_ward)}")

  def totalPatients(self):
    print(f"The number of patients is {len(self.general_ward)+len(self.oxygen_support)}")

  def patientInfo(self):
    print("Patients in oxygen support")
    print("========================================================")
    for patient in self.oxygen_support:
      print("========================================================")
      print("Name:", patient.name)
      print("Admission Date:", patient.admission_date)
      print("symptoms:", patient.symptoms)
      print("oxygen level:", patient.oxy_level)
      print("Discharge Date:", patient.discharge_date)
      print("========================================================")

    print("Patients in General ward")
    print("========================================================")
    for patient in self.general_ward:
      print("========================================================")
      print("Name:", patient.name)
      print("Admission Date:", patient.admission_date)
      print("symptoms:", patient.symptoms)
      print("oxygen level:", patient.oxy_level)
      print("Discharge Date:", patient.discharge_date)
      print("========================================================")

p1 = Patient("suresh", "15/12/2009", ["Dengu", "Nausea"], 50, "25/12/2009")
p2 = Patient("suresh", "02/12/2012", ["Fatigue", "Fever"], 60, "28/12/2012")

db = PatientData()
db.addPatient(p1)
db.addPatient(p2)
db.patientInfo()
db.numberOxygen()
db.numberGeneral()
db.totalPatients()
```

6-Create a Class time that has separate int member data for hours, minutes, and
seconds. One
constructor should initialize this data to 0 and another should initialize it to fixed
values. Another member function should display it in hh:mm: ss format.

```python

class Time:
  def __init__(self, hour = 0, mins = 0, secs = 0):
    self.mins = secs//60
    self.secs = secs%60
    self.mins += mins
    self.hour = self.mins//60
    self.mins = self.mins%60
    self.hour = self.hour%24

  def disp(self):
    print("{0:2d}:{1:2d}:{2:2d}".format(self.hour, self.mins, self.secs))

zero_time = Time()
zero_time.disp()

random_time = Time(25, 67, 120)
random_time.disp()
```
